In this project, the analysis focused on the Iris flower dataset obtained from [source]. The objective was to explore and classify Iris flower species based on their sepal length, sepal width, petal length, and petal width. 

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

### Exploratory Data Analysis

```python
sns.set_style('whitegrid')
sns.scatterplot(data =iris ,x = "SepalLengthCm", y = "SepalWidthCm",hue = "Species",style="Species")
plt.title("Sepal Length vs Width")
plt.show()

sns.lineplot(x="SepalLengthCm", y="SepalWidthCm", hue="Species", data=iris,)
plt.title(" Sepal Length vs Width")
plt.show()
```

### A scatter plot & Line Plot showing the relationship between the sepal width and length for each iris flower specie
- The figures above show that Iris setosa has sepal length not greater than 6.0 cm and sepal width not less than 2.3 cm;
- Iris versicolor has sepal length not less than 5.0 cm and sepal width not greater than 3.4 cm;
- Iris virginica has sepal length not less than 6.0 cm and sepal width not greater than 3.8 cm. 
- The sepal length and width of Iris setosa is clearly distinct from the other species of Iris flower.


```python
sns.scatterplot(data =iris ,x = "PetalLengthCm", y = "PetalWidthCm",hue ="Species",style="Species")
plt.title(" Petal Length vs Width")
plt.show()

```

```python
sns.lineplot(x="PetalLengthCm", y="PetalWidthCm", hue="Species", data=iris)
plt.title(" Petal Length vs Width")
plt.show()
```

# A scatter plot and  Line Plot showing the relationship between the Petal width and length for each iris flower specie
- From the figures above, the petal length and width of Iris setosa is again clearly distinct from the others.
- Iris setosa has petal width and length not greater than 0.6 cm and 2 cm respectively
- Iris versicolor has petal width and petal length not less than 0.9 cm and 3.5 cm respectively
- Iris virginica has petal width and petal length not less than 1.3 cm and 4 cm respectively.

```python
fig, axis = plt.subplots(2, 2, figsize=(11,6))
sns.boxplot(x='Species', y='SepalLengthCm', data=iris, ax=axis[0,0])
sns.boxplot(x='Species', y='SepalWidthCm', data=iris, ax=axis[0,1])
sns.boxplot(x='Species', y='PetalLengthCm', data=iris, ax=axis[1,0])
sns.boxplot(x='Species', y='PetalWidthCm', data=iris, ax=axis[1,1])
plt.show()
```

### This box plot provides valuable insights into the feature variations among the three iris species, aiding in species classification and understanding their distinct characteristics.
### Sepal Length:
- The sepal length of Iris-setosa (blue box) is the shortest, with a median value around 5.0 cm.
- Iris-versicolor (orange box) and Iris-virginica (green box) have similar median sepal lengths, around 5.9 cm.
- There are a few outliers in the sepal length of Iris-virginica.
### Sepal Width:
- Iris-setosa has the widest sepal width, with a median value around 3.4 cm.
- Iris-versicolor and Iris-virginica exhibit similar sepal width distributions, centered around 2.8 cm.
- Some outliers exist in both Iris-versicolor and Iris-virginica.
### Petal Length:
- Iris-setosa has the shortest petal length, with a median value around 1.5 cm.
- Iris-versicolor and Iris-virginica share similar median petal lengths, approximately 4.4 cm.
- No significant outliers are observed.
### Petal Width:
- Iris-setosa has the narrowest petal width, with a median value around 0.2 cm.
- Iris-versicolor and Iris-virginica exhibit similar petal width distributions, centered around 1.3 cm.
- A few outliers are present in both species.
